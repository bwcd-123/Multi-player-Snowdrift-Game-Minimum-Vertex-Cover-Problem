导入软件包、设置超参数

无向图的节点创建、边创建数据见`utils.py`

```python
import os
import matplotlib.pyplot as plt
from PIL import Image

from utils import init_graph, run_game, get_graph


# 设置超参数，默认节点数量为10
r_list = [0.01, 0.05, 0.1, 0.3, 0.5, 0.7, 0.9]
max_epoch = 10 # 最大迭代次数

# 创建保存路径
os.makedirs('figure', exist_ok=True)

# 初始化图
graph, pos = init_graph()
init_image = get_graph(graph, pos)
```

计算每个r值情况下无向图的最终状态

```python
rets, mncs = run_game(graph, pos, r_list, max_epoch)
```

展示初始图片以及各个r值的图片

```python
plt.figure(figsize=(20, 8))
plt.subplot(2, 4, 1)
plt.imshow(Image.open(init_image))
plt.title('initial graph')

for i in range(len(rets)):
    plt.subplot(2, 4, i+2)
    plt.imshow(Image.open(rets[i]))
    plt.title(f'r={r_list[i]}, MNC={mncs[i]}')

plt.tight_layout()
plt.savefig('result.png')
plt.show()
```
